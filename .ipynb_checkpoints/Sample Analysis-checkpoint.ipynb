```python
import pandas as pd
df = pd.read_csv('data_sample/performance.csv', index_col='Time')
df
```

```python
from IPython.core.display import display, HTML
from datetime import datetime, timedelta
def get_duration(start, end):
    t = tuple(int(x) for x in start.split(':'))
    t0 = datetime(1, 1, 1, t[0], t[1], t[2])
    t = tuple(int(x) for x in end.split(':'))
    t1 = datetime(1, 1, 1, t[0], t[1], t[2])
    return str(timedelta(seconds=(t1-t0).total_seconds()))

start = df.index[0]
end = df.index[-1]
duration = get_duration(start, end)
max_users_at = df.loc[df['User Count'] == 10000].index[0]

display(HTML('<h3><u>General Information</u></h3>'))
display(HTML(f'<p><strong>Duration :</strong>  {duration}</p>'))
display(HTML(f'<p><strong>Start time :</strong>  {start}</p>'))
display(HTML(f'<p><strong>End time :</strong>  {end}</p>'))
display(HTML(f'<p><strong>Max Users reached at :</strong>  {max_users_at}</p>'))
display(HTML(f'<p><strong>Statistics :</strong></p>'))
```

<p><strong>Requests/sec Rate</strong></p>
<p>We can notice on the HTML report that the requests/sec rate keeps increasing consistently until 18:31:58, and starts to vacillate after that. The mean Requests/s rate for the test is 407.49, but it appears that this value is soon exceeded at 18:42:49.</p>
<p>This is probably related to my computer's limitations, and we would get much more accurate results with a distributed test..</p>

<p><strong>Response Time</strong></p>
<p>The mean response time for the test is 5.402 seconds, which is quite slow. We can see that despite a Request/sec rate that stays around 407.49, the response time keeps increasing.</p>
<p>We can also notice on the HTML report that the response time seems to peek every 20 minutes starting from 18:40.</p>
<p>An endurance test is recommended.</p>
<p>It is unlinkely but we could also try a spike test to see if quick bursts of user load can induce latency issues.</p>
<p><strong>%Failure</strong><p>
<p>The first failure appears at 18:39:31, with a load of 1638 users, while the mean failure percentage for the test is 36.84%.</p>
<p>We could try a spike test to see how the server reacts to quick bursts of user load, and induce failures before reaching 1638 users.</p>
