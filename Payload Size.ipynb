<p><strong>We can see that the payload size (the request body size to be more precise) can have a huge impact on response size and CPU usage.</strong></p>
<p>The last payload size tested is 134 217 728 bytes (29 iterations). This is because after that, the requests time out (testing with 30sec timeout).</p>
<p>Note that this is not specific to the endpoint, as the homepage presents the same behavior.</p>
<p>Response time seems quite constent after multiple tests.</p>
<p>CPU usage on the other hand is less consistant, but still very sensible to the payload size, and seems to depend on other factors such as how long has it beem up and if it is handling other requests or not.</p>
<p>The results are alarming especially condidering that the test only uses one thread. This could be problematic in case of a DDOS (Distributed Denial Of Service) attempt.</p>
